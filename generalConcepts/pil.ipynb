## Plotting a polygon

```python
from matplotlib.pyplot import imshow
from PIL import Image,ImageDraw
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def get_bbox_image_and_mask(path):

    image=Image.open(path)

    # creating drawable object 
    draw=ImageDraw.Draw(image)


    bbox=[(1100,900),(1925,900),(1925,1300),(1100,1300)]

    polygon=[]
    for i in range(0,len(bbox)):
        polygon.append(bbox[i])

    draw.polygon(polygon,outline='red',width=50)

    mask_image=Image.new('1',image.size,color=0)
    mask_draw=ImageDraw.Draw(mask_image)

    mask_draw.polygon(polygon,fill=1)

    return image,mask_image
```

```python
path='image.png'
bbox_image,mask_img=get_bbox_image_and_mask(path)

# Plot both images side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display bbox_image in the first subplot
axes[0].imshow(bbox_image)
axes[0].set_title("Bounding Box Image")
axes[0].axis('off')  


# Display mask_img in the second subplot
axes[1].imshow(mask_img, cmap='gray')
axes[1].set_title("Mask Image")
axes[1].axis('off')  # Turn off the axis

# Show the plot
plt.tight_layout()
plt.show()


```

## Pixel width

This function calculates the width of the text.

```python
from PIL import Image, ImageFont, ImageDraw, ImageOps

def get_width(font_path, text):
  
    font = ImageFont.truetype(font_path, 24)
    width=font.getlength(text)
    return width

font_path='/home/adi_techbuddy/Desktop/python/repos/brush-your-text/controlnet_util/Textgen/English/Roboto/static/Roboto-Regular.ttf'
caption='I'
get_width(font_path,caption)
```
